#### Are all launch sites in proximity to the Equator line?

The equator is located at latitude 0°. When observing the latitude of the launch sites:
- **CCAFS SLC 40 (28.5623° N)** and **KSC LC 39A (28.573255° N)** are located in Florida, relatively close to the equator (28° North), which is beneficial for launching rockets as it provides extra velocity due to the Earth's rotation.
- **VAFB SLC 4E (34.6321° N)** is located in California, further north from the equator.

Therefore, **not all launch sites are extremely close to the equator**, but they are located in latitudes that are advantageous for space launches, especially the Florida-based sites.

#### Are all launch sites in very close proximity to the coast?

Yes, **all launch sites are near coastal areas**. This is essential for space launches because launching over the ocean provides a safety zone for potential failures and reduces the risk to populated areas.

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```
